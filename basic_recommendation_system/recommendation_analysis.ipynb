```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
credits_df = pd.read_csv("tmdb_5000_credits.csv")
movie_df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
# Analysis of shape of data
print("credits: ", credits_df.shape)
print("movies: ", movie_df.shape)
```

```python
credits_df.rename(columns={"movie_id":"id"},inplace=True)
credits_df.columns
```

```python
merged_df = credits_df.merge(movie_df, on=["id","title"])
merged_df.head()
```

```python
print(merged_df.shape)
```

```python
movies_cleaned_df = merged_df.drop(columns=["title", "status", "homepage", "production_countries"])
movies_cleaned_df.head()
```

```python
movies_cleaned_df.info()
```

## Using weighted average

# ![title](imdb_weightage_formula.png)

```python
v = movies_cleaned_df["vote_count"]
r = movies_cleaned_df["vote_average"]
m = movies_cleaned_df["vote_count"].quantile(0.70)
c = movies_cleaned_df["vote_average"].mean()
```

```python
movies_cleaned_df["weighted_avg"] = (r*v+c*m)/(v+m)
movies_cleaned_df.head()
```

```python
movie_ranking = movies_cleaned_df.sort_values("weighted_avg",ascending=False).reset_index(drop=True)
movie_ranking.head()
```

```python
scaling = MinMaxScaler()
movie_ranking[["normalised_weighted_avg", "normalised_popularity"]] = pd.DataFrame(scaling.fit_transform(movie_ranking[["weighted_avg","popularity"]]),columns=["normalised_weighted_avg", "normalised_popularity"])
movie_ranking.head()
```

```python
movie_ranking["score"] = (movie_ranking["normalised_weighted_avg"]+movie_ranking["normalised_popularity"])/2
movie_ranking = movie_ranking.sort_values("score",ascending=False).reset_index(drop=True)
movie_ranking.head()
```

```python
final_df = movie_ranking[["id", "original_title", "normalised_popularity","normalised_weighted_avg","score"]]
final_df
```
